# Video Analysis


```python
import numpy as np
import cv2
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
model = load_model('animal.h5')
video = cv2.VideoCapture(0)
index=["bears","crows","elephants","rats"]
while 1:
    success,frame=video.read()
    cv2.imwrite('namibia_1057544__340.jpg',frame)
    img=image.load_img('namibia_1057544__340.jpg',target_size=(64,64))
    x=image.img_to_array(img)
    x=np.expand_dims(x,axis=0)
    pred=np.argmax(model.predict(x),axis=1)
    y=pred[0]
    cv2.putText(frame,'The Predicted animal is: '+str(index[y]),(100,100),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),4)
    cv2.imshow('image',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
